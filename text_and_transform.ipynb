```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
```

```python
model_name = "t5-base"  
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python

# Create a pipeline for paraphrasing
paraphraser = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

def improve_sentence(text):
    input_text = f"paraphrase: {text} </s>"  
    output = paraphraser(input_text, max_length=100, do_sample=True, top_k=120, top_p=0.95, num_return_sequences=1)
    return output[0]['generated_text']

```

```python

raw_text = "the movie was good but there some scenes were very boring  unnecessary."
improved_text = improve_sentence(raw_text)
```

```python
print("Original:\n", raw_text)
print("\nImproved Sentence:\n", improved_text)
```
